```python
# filling null values with median as it is more statistically accurate 
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# renameing columns according to snake casting for better readablity and documentation 
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df.rename(columns={'purchase_amount(usd)':'purchase_amount'})
```

```python
# create a new column age group 
labels = ['young_adult','adult','middle_aged','senior']
df['age_group']= pd.qcut(df['age'], q=4,labels=labels)
```

```python
# now to review data 
df[['age','age_group']].head(10)
```

```python
# create new column purchase frequency days 
frequency_mapping = {
    'forenightly':7,
    'weekly':7,
    'monthly':30,
    'quaterly':90,
    'bi_weekly':14,
    'annually':365,
    'every_3_months':90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
# to check whether discount applied and promo code used are same columns 
(df['promo_code_used'] == df['discount_applied']).all
```

```python
# dropping promo code used column 
df = df.drop('promo_code_used', axis=1)
```
